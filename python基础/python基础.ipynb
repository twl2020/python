## 变量
```python
# 变量
a = 10
b = 20
c = a + b
print(c)
```
- 变量名：由字母、数字、下划线组成的序列，且不能以数字开头。
- 变量类型：变量可以存储不同类型的数据，包括整数、浮点数、字符串、布尔值、列表、元组、字典、集合。
- 变量作用域：变量的作用域决定了变量的生命周期，在程序运行期间，变量的作用域可以是全局作用域、局部作用域或嵌套作用域。
- 变量命名规范：
   - 变量名：小写字母、下划线连接
   - 常量名：全部大写字母、下划线连接
- 变量赋值：
   - 直接赋值：将一个值赋给一个变量，如 a = 10。
   - 表达式赋值：将一个表达式的计算结果赋给一个变量，如 a = 10 + 20。
   - 多重赋值：将多个变量的值同时赋值，如 a, b = 10, 20。
- 变量删除：del 语句可以删除一个变量，如 del a。
- 变量作用域：
   - 全局作用域：全局变量可以被所有的函数、模块和类访问，全局变量的生命周期从定义开始，直到程序结束。
   - 局部作用域：局部变量只能被当前函数、模块或类访问，局部变量的生命周期从函数或模块被调用开始，直到函数或模块返回或程序结束。
   - 嵌套作用域：嵌套作用域是指一个作用域中包含另一个作用域，如函数内部定义的函数。
      - 内层作用域可以访问外层作用域的变量，但外层作用域不能访问内层作用域的变量。
      - 内层作用域可以修改外层作用域的变量，但外层作用域不能修改内层作用域的变量。
   - 作用域查找顺序：
      - 局部作用域
      - 内层作用域（如果有）
      - 全局作用域
      - 内层作用域（如果有）
   - global 语句：使用 global 语句可以在函数或模块中声明全局变量，使其可以在函数或模块的局部作用域中访问。
      - 注意：global 语句只能在函数或模块的局部作用域中使用。
            - 内层作用域不能使用 global 语句声明全局变量。
            - 外层作用域不能使用 global 语句声明局部变量。
      - 示例：
         ```python
         x = 10
         def outer():
             global x
             x = 20
             print(x)
         outer()
         print(x)
         ```
         输出：
         ```python
         20
         20
         ```
         说明：
         - 第 1 行：声明全局变量 x。
         - 第 3 行：声明函数 outer，函数内部修改全局变量 x。
         - 第 4 行：使用 global 语句声明变量 x，使其可以在 outer 函数中访问。
         - 第 5 行：调用 outer 函数。
         - 第 6 行：打印变量 x 的值，输出为 20。
         - 第 7 行：打印变量 x 的值，输出为 20。
         - 说明：
            - 变量 x 的值在 outer 函数中被修改为 20，并在函数返回后，被修改为 20。
            - 由于变量 x 的作用域是全局作用域，所以在 outer 函数中修改变量 x，会影响全局变量 x。
            - 由于变量 x 的作用域是全局作用域，所以在 outer 函数中声明变量 x，会使其成为全局变量。
            - 由于变量 x 的作用域是全局作用域，所以在 outer 函数中声明变量 x，会使其可以在函数的局部作用域中访问。
            - 由于变量 x 的作用域是全局作用域，所以在 outer 函数中声明变量 x，会使其可以在函数的内层作用域中访问。
            - 由于变量 x 的作用域是全局作用域，所以在 outer 函数中声明变量 x，会使其可以在函数的外层作用域中访问。
   - nonlocal 语句：使用 nonlocal 语句可以在函数或模块中声明外层作用域的变量，使其可以在函数或模块的嵌套作用域中访问。
      - 注意：nonlocal 语句只能在函数或模块的嵌套作用域中使用。
            - 内层作用域不能使用 nonlocal 语句声明外层作用域的变量。
            - 外层作用域不能使用 nonlocal 语句声明内层作用域的变量。
      - 示例：
         ```python
         def outer():
             y = 20
             def inner():
                 nonlocal y
                 y = 30
                 print(y)
             inner()
             print(y)
         outer()
         ```
         输出：
         ```python
         30
         20
         ```
         说明：
         - 第 1 行：声明全局变量 x。
         - 第 3 行：声明函数 outer，函数内部定义函数 inner。
         - 第 4 行：声明局部变量 y，并赋值为 20。
         - 第 6 行：使用 nonlocal 语句声明变量 y，使其可以在 inner 函数中访问。
         - 第 7 行：修改变量 y 的值为 30。
         - 第 8 行：调用 inner 函数。
         - 第 9 行：打印变量 y 的值，输出为 30。
         - 第 10 行：打印变量 y 的值，输出为 20。
         - 第 11 行：调用 outer 函数。
         - 第 12 行：打印变量 y 的值，输出为 20。
         - 说明：
            - 变量 y 的值在 inner 函数中被修改为 30，并在函数返回后，被修改为 20。
            - 变量 y 的值在 outer 函数中被修改为 20，并在函数返回后，被修改为 20。
            - 由于变量 y 的作用域是局部作用域，所以在 inner 函数中修改变量 y，不会影响外层作用域的变量 y。
            - 由于变量 y 的作用域是局部作用域，所以在 outer 函数中修改变量 y，不会影响 inner 函数的变量 y。
            - 由于变量 y 的作用域是嵌套作用域，所以 inner 函数可以访问外层作用域的变量 y。
            - 由于变量 y 的作用域是嵌套作用域，所以 outer 函数可以访问 inner 函数的变量 y。
    - locals() 函数：locals() 函数可以获取当前作用域中的所有变量。
    - globals() 函数：globals() 函数可以获取全局作用域中的所有变量。
        - 注意：globals() 函数只能在模块的全局作用域中使用。
                - 不能在函数或类中使用 globals() 函数。
                        - 示例：
                            ```python
                            x = 10
                            def outer():
                                y = 20
                                def inner():
                                    print(globals())
                                inner()
                            outer()
                            ```
                            输出：
                            ```python
                            {'__name__': '__main__', 'x': 10, 'outer': <function outer at 0x000001B7F2F0D1F0>, 'inner': <function inner at 0x000001B7F2F0D288>}
                            ```
                            说明：
                            - 第 1 行：声明全局变量 x。
                            - 第 3 行：声明函数 outer，函数内部定义函数 inner。
                            - 第 4 行：声明局部变量 y，并赋值为 20。
                            - 第 6 行：调用 inner 函数。
                            - 第 7 行：打印 globals() 函数的返回值，输出为字典。
                            - 说明：
                                - globals() 函数返回当前作用域中的所有变量，包括全局变量和局部变量。
                                - 由于 inner 函数是在 outer 函数中定义的，所以 inner 函数的作用域是局部作用域。
                                - 由于 inner 函数的作用域是局部作用域，所以 globals() 函数只能在 inner 函数中使用。
                                - 由于 inner 函数的作用域是局部作用域，所以 globals() 函数返回的字典中只包含 inner 函数的局部变量。
                                - 由于 outer 函数是在全局作用域中定义的，所以 outer 函数的作用域是全局作用域。
                                - 由于 outer 函数的作用域是全局作用域，所以 globals() 函数返回的字典中包含全局变量 x。



## 条件语句
```python
# 条件语句
a = 10
b = 20
if a > b:
    print("a is greater than b")
elif a == b:
    print("a is equal to b")
else:
    print("a is less than b")
```

- 默认参数：默认参数用于设置参数的默认值，语法格式为：
   - def 函数名(参数列表, 默认参数=默认值)
- 可变参位置数：可变参数用于将参数个数不定的参数传递给函数，语法格式为：
   - def 函数名(参数列表, *args)
   - 示例：
      ```python
      def my_sum(*args):
          result = 0
          for i in args:
              result += i
          return result
      print(my_sum(1, 2, 3, 4, 5))
      ```
      输出：
      ```python
      15
      ```
      说明：
      - 第 1 行：定义函数 my_sum，参数列表中包含可变参数 args。
      - 第 3 行：定义变量 result，并初始化为 0。
      - 第 4 行：遍历可变参数 args，将每个元素的值加到变量 result 中。
      - 第 5 行：返回变量 result 的值。
      - 第 6 行：调用函数 my_sum，传入参数 1、2、3、4、5。
      - 第 7 行：打印函数 my_sum 的返回值，输出为 15。
- 可变关键字参数：关键字参数用于将参数按照名称传递给函数，语法格式为：
   - def 函数名(参数列表, **kwargs)

      ```python
      def my_func(a, b, c=10, d=20):
          print("a =", a)
          print("b =", b)
          print("c =", c)
          print("d =", d)
      my_func(10, 20, c=30, d=40)
      ```
      输出：
      ```python

      b = 20
      c = 30
      d = 40
      ```
      说明：
      - 第 1 行：定义函数 my_func，参数列表中包含关键字参数 c 和 d，默认值为 10 和 20。
      - 第 4 行：打印函数 my_func 的参数 a、b、c、d 的值。
      - 第 5 行：调用函数 my_func，传入参数 a=10、b=20、c=30、d=40。

## 异常处理
```python
# 异常处理
try:
    a = 10 / 0
except ZeroDivisionError:
    print("division by zero!")
except:
    print("something else went wrong")
else:
    print("no exceptions were raised")
finally:
    print("cleaning up, regardless of whether an exception was raised or not")
```
- try 语句：try 语句用于捕获异常，语法格式为：
   - try:
      - 可能引发异常的代码
   - except 异常类型1:
      - 异常类型1发生时执行的代码
   - except 异常类型2:
      - 异常类型2发生时执行的代码
   - ...
   - except 异常类型n:
      - 异常类型n发生时执行的代码
   - else:
      - 没有异常发生时执行的代码
   - finally:
      - 无论是否发生异常，都会执行的代码
- 异常类型：
   - BaseException：所有异常的基类。
   - Exception：常规异常，包括系统错误、断言错误等。

   - 线程：线程是操作系统调度的最小单位，它是CPU调度的基本单位，一个线程只能属于一个进程。
   - 进程和线程的区别：
      - 进程：
         - 进程是资源分配的基本单位，它是系统运行程序的基本单位，是动态创建的。
         - 进程可以创建新的进程，创建进程的开销比线程小。
         - 进程间通信比较复杂，需要IPC（Inter-Process Communication，进程间通信）。
         - 进程可以执行多线程，但线程比进程更加轻量级。
      - 线程：
         - 线程可以执行多次，但进程只能执行一次。
         - 线程间通信比较简单，不需要IPC。
         - 线程比进程更加轻量级，创建线程的开销比进程小。
  - 建议：
     - 对于CPU密集型任务，使用多进程。
     - 对于IO密集型任务，使用多线程。
  - 示例：
      ```python
      import multiprocessing
      import threading

      def my_func(name):
          print("hello, " + name)

      if __name__ == '__main__':
          # 创建进程池
          pool = multiprocessing.Pool(processes=4)
          # 异步执行任务
          results = [pool.apply_async(my_func, args=("world",)) for i in range(10)]
          # 等待所有任务结束
          pool.close()
          pool.join()
          # 打印结果
          for result in results:
              print(result.get())

          # 创建线程池
          pool = threading.Pool(processes=4)
          # 异步执行任务
          results = [pool.apply_async(my_func, args=("world",)) for i in range(10)]
          # 等待所有任务结束
      ```

## 魔术方法
- 魔术方法（Magic）：在Python中，以双下划线开头和结尾的特殊方法，称为魔术方法。
- 魔术方法的作用：
- 魔术方法的分类：
  - 实例方法：以self作为第一个参数的方法，实例方法可以访问和修改类的实例变量。
  - 类方法：以cls作为第一个参数的方法，类方法可以访问和修改类变量。
  - 静态方法：以@staticmethod装饰的方法，静态方法不访问或修改实例或类变量。
- 魔术方法的使用：
  - 实例方法：
    - __init__：构造方法，在创建对象时调用。
    - __del__：析构方法，在对象被删除时调用。
    - __str__：打印对象时调用。
    - __repr__：调试时调用。
    - __getattr__：访问不存在的属性时调用。
    - __setattr__：设置不存在的属性时调用。
    - __getattribute__：访问属性时调用。
    - __call__：对象作为函数调用时调用。
  - 类方法：
    - __new__：创建对象时调用。
    - __init_subclass__：子类创建时调用。
    - __prepare__：子类创建时调用。
  - 静态方法：
    - @staticmethod：装饰器，将方法转换为静态方法。
- 示例：
  ```python
  class MyClass:
      def __init__(self, name):
          self.name = name

      def __str__(self):
          return "MyClass object (name: %s)" % self.name

      def __call__(self, *args, **kwargs):
          print("Calling object with args: %s, kwargs: %s" % (args, kwargs))

  obj = MyClass("Alice")
  print(obj)
  obj("hello", "world")
  ```
  输出：
  ```python
  MyClass object (name: Alice)
  Calling object with args: ('hello', 'world'), kwargs: {}
  ```
  说明：
  - 第 1 行：定义类 MyClass。
  - 第 4 行：定义构造方法 __init__。
  - 第 7 行：定义打印方法 __str__。
  - 第 10 行：定义调用方法 __call__。
  - 第 13 行：创建对象 obj。
  - 第 14 行：打印对象 obj。
  - 第 15 行：调用对象 obj。
注意：
  - 实例方法：
    - self：实例对象。
    - self.name：实例变量。
  - 类方法：
    - cls：类对象。
    - cls.name：类变量。
  - 静态方法：
    - 无需实例或类对象。
