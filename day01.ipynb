## python基础 (https://docs.python.org/zh-cn/3.11/tutorial/introduction.html)
```shell
1.本语言的命名源自 BBC 的 “Monty Python 飞行马戏团”，与爬行动物无关（Python 原义为“蟒蛇”）
2.python的解释器 解释器的操作方式类似 Unix Shell：用与 tty 设备关联的标准输入调用时，
  可以交互式地读取和执行命令；以文件名参数，或标准输入文件调用时，则读取并执行文件中的 脚本
3.另一种启动解释器的方式是 python -c command [arg] ...，这将执行 command 中的语句，相当于 shell 的 -c 选项。  python -c "print(min(1,2))"
  由于 Python 语句经常包含空格或其他会被 shell 特殊对待的字符，通常建议用引号   将整个 command 括起来。
  Python 模块也可以当作脚本使用。输入：python -m module [arg] ...，会执行 module 的源文件，这跟在命令行把路径写全了一样。  python -m main
  在交互模式下运行脚本文件，只要在脚本名称参数前，加上选项 -i 就可以了。  python -i main.py
```

```shell

1.Python 程序的基本组成部分是模块（module）、函数（function）、类（class）和语句（statement）。模块是 Python 代码的基本单位，包含了 Python 代码定义和执行的逻辑。模块可以被导入（import）、使用（使用 from 模块名 import 成员名 导入模块中的成员）、重新定义（使用 as 给模块取别名）、修改（使用 reload() 函数重新加载模块）、执行（使用 exec() 函数执行模块中的代码）。模块的名称通常是文件名（不带 .py 扩展名）。

2.函数是模块的组成部分，它定义了一个输入（参数）和输出（返回值）的计算过程。函数可以被调用（调用时传入参数）、定义（定义时定义参数和返回值）、修改（使用 def 语句重新定义）、删除（使用 del 语句删除）。函数的名称通常是小写字母，多个单词用下划线连接。

3.类是模块的组成部分，它定义了对象的属性和行为。类可以被定义（使用 class 语句定义）、实例化（使用类名() 创建对象）、修改（使用属性赋值语句修改对象属性）、删除（使用 del 语句删除对象）。类名通常是大写字母，多个单词用下划线连接。

4.语句是 Python 代码的基本组成部分，它是一条或多条 Python 代码的集合。语句可以是表达式（表达式的值是计算结果）、赋值语句（将值赋给变量）、条件语句（根据条件执行不同的代码块）、循环语句（重复执行代码块）、打印语句（输出值到屏幕或文件）、输入语句（从键盘输入值）、函数调用语句（调用函数）、类实例化语句（创建类的实例）、异常处理语句（处理异常）。

5.Python 程序的执行顺序是从上到下，从左到右。程序的执行可以暂停并等待用户输入，也可以在程序执行过程中随时中断。程序的执行可以被中断，并在下次继续执行。

6.Python 程序的语法遵循“缩进”规则，即代码块（如函数或类定义）的第一行必须缩进。缩进的空格数是可变的，但必须相同。

7.Python 程序的注释以 # 开头，注释可以单独占一行，也可以与代码混合在一起。

8.Python 程序的变量名必须是有效的标识符，即由字母、数字、下划线组成的序列，且不能以数字开头。

9.Python 程序的标识符的命名规范：
   - 变量名：小写字母、下划线连接
   - 函数名：小写字母、下划线连接
   - 类名：大写字母、下划线连接
   - 模块名：小写字母、下划线连接
   - 常量名：全部大写字母、下划线连接

10.Python 程序的运算符有：
   - 算术运算符：+、-、*、/、//、%、**
   - 赋值运算符：=、+=、-=、*=、/=、%=、**=
   - 比较运算符：==、!=、>、<、>=、<=
   - 逻辑运算符：and、or、not
   - 位运算符：&、|、^、~、<<、>>
   - 成员运算符：in、not in
   - 身份运算符：is、is not

11.Python 程序的控制流语句有：
   - if 语句
   - for 语句
   - while 语句
   - try-except 语句
   - with 语句

12.Python 程序的输入输出有：
   - 打印输出：print() 函数
   - 输入输出：input() 函数
   - 文件输入输出：open() 函数

13.Python 程序的异常处理有：
   - try-except 语句
   - raise 语句
   - assert 语句

14.Python 程序的调试有：
   - pdb 调试器
   - 单元测试

15.Python 程序的编码风格有：
   - PEP 8 编码风格指南
   - Google Python 风格指南 (https://zh-google-styleguide.readthedocs.io/en/latest/google-python-styleguide/contents/)
   - 其他编码风格指南
16.Python 程序的运行环境有：
   - Python 解释器
   - Python 标准库
   - 第三方库
   - 虚拟环境
   - 集成开发环境（IDE）
```

数据类型详解：

- 浮点数（float）：浮点数类型用于存储小数值，包括正数、负数、零、无限大。
- 字符串（str）：字符串类型用于存储文本数据，包括单个字符和多个字符。
- 布尔值（bool）：布尔值类型用于存储布尔值，包括 True 和 False。
- 列表（list）：列表类型用于存储一系列值，这些值可以是不同类型的数据。列表可以包含重复的值。
- 元组（tuple）：元组类型用于存储一系列值，这些值可以是不同类型的数据。元组中的值不能被修改。
- 字典（dict）：字典类型用于存储一系列键值对，其中每个键都对应一个值。字典中的键必须是唯一的。
- 集合（set）：集合类型用于存储一系列值，这些值不能重复。集合中的值可以是不同类型的数据。
- 类型转换：可以将不同类型的数据转换为相同类型的数据。


## 变量
```python
# 变量
a = 10
b = 20
c = a + b
print(c)
```
- 变量名：由字母、数字、下划线组成的序列，且不能以数字开头。
- 变量类型：变量可以存储不同类型的数据，包括整数、浮点数、字符串、布尔值、列表、元组、字典、集合。
- 变量作用域：变量的作用域决定了变量的生命周期，在程序运行期间，变量的作用域可以是全局作用域、局部作用域或嵌套作用域。
- 变量命名规范：
   - 变量名：小写字母、下划线连接
   - 常量名：全部大写字母、下划线连接
- 变量赋值：
   - 直接赋值：将一个值赋给一个变量，如 a = 10。
   - 表达式赋值：将一个表达式的计算结果赋给一个变量，如 a = 10 + 20。
   - 多重赋值：将多个变量的值同时赋值，如 a, b = 10, 20。
- 变量删除：del 语句可以删除一个变量，如 del a。
- 变量作用域：
   - 全局作用域：全局变量可以被所有的函数、模块和类访问，全局变量的生命周期从定义开始，直到程序结束。
   - 局部作用域：局部变量只能被当前函数、模块或类访问，局部变量的生命周期从函数或模块被调用开始，直到函数或模块返回或程序结束。
   - 嵌套作用域：嵌套作用域是指一个作用域中包含另一个作用域，如函数内部定义的函数。
      - 内层作用域可以访问外层作用域的变量，但外层作用域不能访问内层作用域的变量。
      - 内层作用域可以修改外层作用域的变量，但外层作用域不能修改内层作用域的变量。
   - 作用域查找顺序：
      - 局部作用域
      - 内层作用域（如果有）
      - 全局作用域
      - 内层作用域（如果有）
   - global 语句：使用 global 语句可以在函数或模块中声明全局变量，使其可以在函数或模块的局部作用域中访问。
   - local 语句：local变量是指在函数或模块的局部作用域中定义的变量。
      - 注意：local 语句只能在函数或模块的局部作用域中使用。
            - 内层作用域不能使用 local 语句声明变量。
            - 外层作用域不能使用 local 语句声明变量。
      - 示例：
         ```python
         x = 10
         def outer():
             y = 20
             def inner():
                 local y
                 y = 30
                 print(y)
             inner()
             print(y)
         outer()
         ```
         输出：
         ```python
         30
   - nonlocal 语句：使用 nonlocal 语句可以在函数或模块中声明外层作用域的变量，使其可以在函数或模块的嵌套作用域中访问。
      - 注意：nonlocal 语句只能在函数或模块的嵌套作用域中使用。
            - 内层作用域不能使用 nonlocal 语句声明外层作用域的变量。
            - 外层作用域不能使用 nonlocal 语句声明内层作用域的变量。
      - 示例：
         ```python
         x = 10
         def outer():
             y = 20
             def inner():
                 nonlocal y
                 y = 30
                 print(y)
             inner()
             print(y)
         outer()
         ```
         输出：
         ```python
         30
         20
         ```
         说明：
         - 第 1 行：声明全局变量 x。
         - 第 3 行：声明函数 outer，函数内部定义函数 inner。
         - 第 4 行：声明局部变量 y，并赋值为 20。
         - 第 6 行：使用 nonlocal 语句声明变量 y，使其可以在 inner 函数中访问。
         - 第 7 行：修改变量 y 的值为 30。
         - 第 8 行：调用 inner 函数。
         - 第 9 行：打印变量 y 的值，输出为 30。
         - 第 10 行：打印变量 y 的值，输出为 20。
         - 第 11 行：调用 outer 函数。
         - 第 12 行：打印变量 y 的值，输出为 20。
         - 说明：
            - 变量 y 的值在 inner 函数中被修改为 30，并在函数返回后，被修改为 20。
            - 变量 y 的值在 outer 函数中被修改为 20，并在函数返回后，被修改为 20。
            - 由于变量 y 的作用域是局部作用域，所以在 inner 函数中修改变量 y，不会影响外层作用域的变量 y。
            - 由于变量 y 的作用域是局部作用域，所以在 outer 函数中修改变量 y，不会影响 inner 函数的变量 y。
            - 由于变量 y 的作用域是嵌套作用域，所以 inner 函数可以访问外层作用域的变量 y。
            - 由于变量 y 的作用域是嵌套作用域，所以 outer 函数可以访问 inner 函数的变量 y。
    - locals() 函数：locals() 函数可以获取当前作用域中的所有变量。
    - globals() 函数：globals() 函数可以获取全局作用域中的所有变量。
        - 注意：globals() 函数只能在模块的全局作用域中使用。
                - 不能在函数或类中使用 globals() 函数。
                        - 示例：
                            ```python
                            x = 10
                            def outer():
                                y = 20
                                def inner():
                                    print(globals())
                                inner()
                            outer()
                            ```
                            输出：
                            ```python
                            {'__name__': '__main__', 'x': 10, 'outer': <function outer at 0x000001B7F2F0D1F0>, 'inner': <function inner at 0x000001B7F2F0D288>}
                            ```
                            说明：
                            - 第 1 行：声明全局变量 x。
                            - 第 3 行：声明函数 outer，函数内部定义函数 inner。
                            - 第 4 行：声明局部变量 y，并赋值为 20。
                            - 第 6 行：调用 inner 函数。
                            - 第 7 行：打印 globals() 函数的返回值，输出为字典。
                            - 说明：
                                - globals() 函数返回当前作用域中的所有变量，包括全局变量和局部变量。
                                - 由于 inner 函数是在 outer 函数中定义的，所以 inner 函数的作用域是局部作用域。
                                - 由于 inner 函数的作用域是局部作用域，所以 globals() 函数只能在 inner 函数中使用。
                                - 由于 inner 函数的作用域是局部作用域，所以 globals() 函数返回的字典中只包含 inner 函数的局部变量。
                                - 由于 outer 函数是在全局作用域中定义的，所以 outer 函数的作用域是全局作用域。
                                - 由于 outer 函数的作用域是全局作用域，所以 globals() 函数返回的字典中包含全局变量 x。
